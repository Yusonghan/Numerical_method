```python
import random
from copy import deepcopy


class Matrix:

    def __init__(self, nrows, ncols):
        """Construct a (nrows X ncols) matrix"""
        self.nrows=nrows # creates an attribute called nrows and assigns to it the value of the nrows parameter. 
        self.ncols=ncols # creates an attribute called ncols and assigns to it the value of the ncols parameter. 
        self.matrix=[]
        for i in range (self.nrows): #隨機值寫入矩陣
            self.matrix.append([])
            for j in range (self.ncols):
                ran=random.randint(1,10)
                self.matrix[i].append(ran)

    def __setitem__(self,idx,item):
        self.matrix[idx]=item   #item是一個陣列     
    
    def add(self, m):
        """return a new Matrix object after summation"""
        # Check two matrix are in the same size
        if self.ncols != m.ncols and self.nrows != m.nrows: #矩陣行數獲列數不相等
            print("Matrixs' size should be in the same size")
            return None
        # add self.matrix with m.matrix
        result = deepcopy(self)
        for x in range(self.nrows):
            row= [sum(item) for item in zip(self.matrix[x],m.matrix[x])] #把兩個矩陣的相同位置鏈起來相加
            result [x]=row #一次加一整列
        return result

    def sub(self, m):
        """return a new Matrix object after substraction"""
        # Check two matrix are in the same size
        if self.ncols != m.ncols and self.nrows != m.nrows: #矩陣行數獲列數不相等
            print("Matrixs' size should be in the same size")
            return None
        # substraction self.matrix from m.matrix
        result = deepcopy(self)
        for x in range(self.nrows):
            row= [item[0]-item[1] for item in zip(self.matrix[x],m.matrix[x])] #把兩個矩陣的相同位置鏈起來相減
            result [x]=row #一次加一整列
        return result

    def mul(self, m):
        """return a new Matrix object after multiplication"""
        # Check if the nrows of m is the same as ncols of self
        if self.ncols != m.nrows: #A的行數不等於B的列數
            print("Cannot multiply two matrixs with size ({},{})and({},{})".format(self.nrows,self.ncols,m.nrows,m.ncols))
            return None
        # Multiply self.matrix and m.matrix
        result = Matrix(self.nrows, m.ncols) 
        m_m=m.transpose() #轉置B矩陣
        for i in range (self.nrows): #每個位置相乘，無法項加減以列運算
            for j in range (self.ncols):
                result.matrix[i][j]=sum([item[0]*item[1] for item in zip(self.matrix[i],m_m.matrix[j])]) #把A矩陣與轉置的B矩陣的相同位置鏈起來相乘  
        return result

    def transpose(self):
        """return a new Matrix object after transpose"""
        # Tranpose
        #if result is None:
         #   return None
        result=Matrix(self.nrows,self.ncols)
        result.matrix=[list(item) for item in zip(*self.matrix)]    
        return result
    
    def display(self):
        """Display the content in the matrix"""
        for i in range (self.nrows):
            print("\n")
            for j in range (self.ncols):
                print(self.matrix[i][j],end=' ')
        print("\n")
```

```python
a_rows = int(input("Enter A matrix's rows:"))
a_cols = int(input("Enter A matrix's cols:"))
print("Matrix A({}, {}):".format(a_rows, a_cols))
A = Matrix(a_rows, a_cols)
A.display()
    
b_rows = int(input("Enter B matrix's rows:"))
b_cols = int(input("Enter B matrix's cols:"))
print("Matrix B({}, {}):".format(b_rows, b_cols))
B = Matrix(b_rows, b_cols)
B.display()
    
print("="*10, 'A + B', "="*10)
result = A.add(B)
if result is not None:
    result.display()
    
print("="*10, 'A - B', "="*10)
result = A.sub(B)
if result is not None:
    result.display()

print("="*10, 'A * B', "="*10)
result = A.mul(B)
if result is not None:
    result.display()

print("="*5, "the transpose of A*B", "="*5)
result = result.transpose()
if result is not None:
    result.display()
```
