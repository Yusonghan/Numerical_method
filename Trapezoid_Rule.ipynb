```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
def function(x):
    f=2*(x**2)+(1/3)*x+12
    return f
```

```python
#區間數量
n=51
#下限
lower_limit=-3
#上限
upper_limit=9
#區間大小
h=(upper_limit-lower_limit)/n
#print(h)
```

```python
result,error=quad(function,lower_limit,upper_limit)
print(result)
```

```python
def Integral_trapezoid(f,a,b,h):
    x=np.arange(a,b+h,h) #len=51
    A=f(x)
    #print(A)
    B=[]
    for i in range(len(f(x))-1):
        B.append(h*(A[i]+A[i+1])/2)
    return B
```

```python
Integral_trapezoid_result=Integral_trapezoid(function,lower_limit,upper_limit,h)
total=sum(Integral_trapezoid_result)
#print(function(upper_limit)) #確認邊界
```

```python
x=np.arange(lower_limit,upper_limit,h)
#print(len(function(x)))
error=result-total
A=function(x)

print('Integral_trapezoid:', sum(Integral_trapezoid_result))
print('error:', error)
plt.plot(x , function(x), linewidth=2, c = 'r')
for i in range(0, n-1):
    c = ['darkorange', 'forestgreen', 'royalblue']
    x_range = [x[i], x[i], x[i+1], x[i+1]]
    y_range = [0, A[i], A[i+1], 0]
    plt.fill(x_range, y_range, color=c[i%3], alpha = 0.5)
plt.show()
```
