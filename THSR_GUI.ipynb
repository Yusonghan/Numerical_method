```python
import numpy as np
import pandas as pd
import csv
import flet as ft
```

```python
# 讀資料

ticket_df = pd.read_csv('tickets.csv', header=None)
ticket = ticket_df.values
ticket_df
```

```python
# 南下的組合

southbound_dict = {}
#print(len(ticket))
for i in range(len(ticket[0])): #[0,12]
    for j in range(i+1, len(ticket[0])-1): #[1,11]
        station_1 = str(ticket[0][i+1])
        station_2 = str(ticket[0][j+1])
        southbound_dict[(station_1, station_2)] = (ticket[j+1][i+1], ticket[i+1][j+1])
```

```python
# 北上的組合

northbound_dict = {}
for i in range(len(ticket[0])): 
    for j in range(i+1, len(ticket[0])-1):
        station_1 = str(ticket[0][-i-1]) #倒過來
        station_2 = str(ticket[0][-j-1])
        northbound_dict[(station_1, station_2)] = (ticket[-i-1][-j-1], ticket[-j-1][-i-1])
```

```python
# 站名

station = []
for i in range(1, len(ticket[0])): #[1,12]
    station.append(str(ticket[0][i]))
#print(station)
```

```python
# 這個function只是給大家感受一下GUI大概會怎麼被使用的樣子。接收到2個input，然後output票價的結果

def THSR_fare():
    start_station = input('Enter your start station:')
    end_station = input('Enter your end station:')
    if southbound_dict.get((start_station, end_station)) is not None:
        print(f'Southbound from {start_station} to {end_station}')
        print(f'The ticket fare of Standard Car is: {southbound_dict[start_station, end_station][0]}')
        print(f'The ticket fare of Business Car is: {southbound_dict[start_station, end_station][1]}')
    elif northbound_dict.get((start_station, end_station)) is not None:
        print(f'Northbound from {start_station} to {end_station}')
        print(f'The ticket fare of Standard Car is: {northbound_dict[start_station, end_station][0]}')
        print(f'The ticket fare of Business Car is: {northbound_dict[start_station, end_station][1]}')
    else:
        print(f'The destination from {start_station} to {end_station} is not found')
```

## 開始GUI介面實作

###### 大家不一定要照著下面的模板實作，只要能夠做出相同功能、相似尺寸的GUI就好

```python
def main(page: ft.Page):
    global start_station, end_station

    # GUI的排版
    page.title = "Taiwan High Speed Rail Fare System"
    page.window_width = 750
    page.window_height = 600
    page.horizontal_alignment = ft.CrossAxisAlignment.CENTER #置中

    start_station = ""
    end_station = ""

    # 按鈕的function
    def start_station_click(aaa):
        # 使用global變數定義start_station讓其他function也能夠使用到被更新後的同樣參數
        # 可以使用aaa.control.data抓到站名
        global start_station
        #for i in range(len(station)):
        #    ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=start_station_click)
        start_station=aaa.control.data #抓站名
        t1.value=f"The start station of your choice:{start_station}" #輸出
        page.update()
    
    def end_station_click(aaa):
        global end_station
        #for i in range(len(station)):
        #    ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=end_station_click)
        end_station=aaa.control.data
        t2.value=f"The end station of your choice:{end_station}"
        page.update()

    
    def result_click(aaa):
        # 可參考THSR_fare()這個function去實作這邊
        if southbound_dict.get((start_station, end_station)) is not None:
            #ft.Text(f'Southbound from {start_station} to {end_station}')
            t3.value=f'The ticket fare of Standard Car is: {southbound_dict[start_station, end_station][0]}\nThe ticket fare of Business Car is: {southbound_dict[start_station, end_station][1]}'
        elif northbound_dict.get((start_station, end_station)) is not None:
            #ft.Text(f'Northbound from {start_station} to {end_station}')
            t3.value=f'The ticket fare of Standard Car is: {northbound_dict[start_station, end_station][0]}\nThe ticket fare of Business Car is: {northbound_dict[start_station, end_station][1]}'
        else:
            t3.value=f'The destination from {start_station} to {end_station} is not found'
        page.update()

    # ------建立物件------
    start_text = ft.Text("Please select a start station", size=18)
    end_text = ft.Text("Please select a end station", size=18)
    
    # /// for start station view /// 
    # 可以一個一個建立出按鈕，但也可以透過for loop 去建立出 4(column)*3(row)的樣子
    # 使用station這個參數可以抓到各個站名
    # 使用這個函數實作，參數可以自己調整，除了width
    # ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=start_station_click)
    item1=[] #第一列
    item2=[] #第二列
    item3=[] #第三列 
    #print(len(station))
    for i in range(0,4):
        item1.append(ft.Container(ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=start_station_click)))
    row1=ft.Row(spacing=35,controls=item1,alignment=ft.MainAxisAlignment.CENTER)
    for i in range(4,8):
        item2.append(ft.Container(ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=start_station_click)))
    row2=ft.Row(spacing=35,controls=item2,alignment=ft.MainAxisAlignment.CENTER)
    for i in range(8,12):
        item3.append(ft.Container(ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=start_station_click)))
    row3=ft.Row(spacing=35,controls=item3,alignment=ft.MainAxisAlignment.CENTER)
    col1=ft.Column(spacing=10,controls=[row1,row2,row3]) #三列合成一行

    # /// for end station view ///
    item4=[] #同上，只是按鈕不同，不能重複使用
    item5=[]
    item6=[]
    #print(len(station))
    for i in range(0,4):
        item4.append(ft.Container(ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=end_station_click)))
    row4=ft.Row(spacing=35,controls=item4,alignment=ft.MainAxisAlignment.CENTER)
    for i in range(4,8):
        item5.append(ft.Container(ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=end_station_click)))
    row5=ft.Row(spacing=35,controls=item5,alignment=ft.MainAxisAlignment.CENTER)
    for i in range(8,12):
        item6.append(ft.Container(ft.ElevatedButton(text=f"{station[i]}", data=f"{station[i]}", width=150, on_click=end_station_click)))
    row6=ft.Row(spacing=35,controls=item6,alignment=ft.MainAxisAlignment.CENTER)
    col2=ft.Column(spacing=10,controls=[row4,row5,row6])
    
    
    # /// for result view ///
    # 使用以下函數實作
    # ft.ElevatedButton(text=f"Calculate the fare", width=630, on_click=result_click)
    # ft.Text("")
    result=ft.ElevatedButton(text=f"Calculate the fare", width=630, on_click=result_click)
    t1=ft.Text() #起始站文字呈現
    t2=ft.Text() #終點站文字呈現
    t3=ft.Text() #價格文字呈現


    # ------將物件進行排版------
    page.add(start_text,col1,end_text,col2,result)
    page.add(t1,t2,t3)

ft.app(target=main)
```
