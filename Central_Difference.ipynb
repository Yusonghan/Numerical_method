```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Central_Difference_Smoothing(x,y,n):
    y_smooth=[]
    i=n
    for i in range(n,len(y)-n):
        y_smooth.append(np.mean(y[i-n:i+n]))
    h=(2*np.pi/100)
    X=np.arange(h,2*np.pi-h,h)
    dy=[]
    for i in range(len(y)-2):   
        dy.append((y[i+2]-y[i])/(2*h))
    return dy,X
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.randn(len(x))/100
#print(x)
[dy, X] = Central_Difference_Smoothing(x, y, n=4)
#print(len(dy)) #確認長度
#print(len(X)) #確認長度
```

```python
#plot solution
plt.figure(figsize=(12,6))
plt.plot(x,y,"b-",label="sin(x)+noise")
plt.title("Noisy sine function")
plt.xlabel("x")
plt.ylabel("y")
#plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)
yy=np.sin(x)
A=[]
h=(2*np.pi/100)
for i in range(len(yy)-2):   
        A.append((yy[i+2]-yy[i])/(2*h))

#plot solution
plt.figure(figsize=(12,6))
plt.plot(x,y,"b-",label="cosine")
plt.plot(X,dy,"g-",label="unsmoothed Central diff")
plt.plot(X,A,"r-",label="smoothed")
plt.title("Analytic Derivative and Smoothed Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
