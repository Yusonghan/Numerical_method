```python
# define parameter for plt
linewidths = [0.5, 1.0, 2.0, 4.0]
linestyles = ['-', '--', '-.', ':']
markers = ['+', 'o', '*', 's', '.', '1', '2', '3', '4']
markersizecolors = [(4, "white"), (8, "red"), (12, "yellow"), (16, "lightgreen")] #tuple in list
```

```python
import numpy as np #make numpy simplify to np
import matplotlib.pyplot as plt #make matplotlib.pyplot simplify to plt
x = np.linspace(-5, 5, 5) 
y = np.ones_like(x)                                        # return an array of 1 with the same shape and type as x. 
def axes_settings(fig, ax, title, ymax):
    ax.set_xticks([])                                      # No xticks
    ax.set_yticks([])                                      # No yticks
    ax.set_ylim(0, ymax+1)
    ax.set_title(title)

fig, axes = plt.subplots(1, 4, figsize=(16,3))             # create a figure with 1 row X 4 columns subfig 
# Line width
for n, linewidth in enumerate(linewidths):                 # enumerate returns index and value
    axes[0].plot(x, y + n, color="blue", linewidth=linewidth)
axes_settings(fig, axes[0], "linewidth", len(linewidths))

# Line style
for n, linestyle in enumerate(linestyles):
    axes[1].plot(x, y + n, color='blue',linestyle=linestyle) #第二位置，linestyle=linstyles陣列
axes_settings(fig,axes[1],'linestyle',len(linestyles)) #標題:linestyle，y_max=4

# Marker
for n, marker in enumerate(markers):
    axes[2].plot(x, y + n, color='blue',marker=marker,linestyle='None') #marker=markers陣列，linestyle=None for marker without line
axes_settings(fig,axes[2],'marker',len(markers)) #標題:marker，y_max=9

# marker size/color
for n, (size, color) in enumerate(markersizecolors): #將markersizecolor分為size和color
    axes[3].plot(x, y + n, "o",markerfacecolor=color,markersize=size,markeredgecolor="blue",markeredgewidth="2") #分別對應size及color，marker外框="2"
axes_settings(fig,axes[3],'markersizecolor',len(markersizecolors)) #標題:markersizecolor，y_max=4
```
